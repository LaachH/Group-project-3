```python
import numpy as np 
import pandas as pd 
import os
from plotly.offline import init_notebook_mode, iplot
import plotly as py
import plotly.figure_factory as FF

from plotly import tools
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
import datetime

from pathlib import Path
warnings.filterwarnings("ignore")
init_notebook_mode(connected=True)
```

```python
melb_stats = Path("MELBOURNE_HOUSE_PRICES_LESS.csv")
melb_stats2 = pd.read_csv(melb_stats, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
melb_stats2
```

```python
# Create the sunburst chart
```

```python
# Create a new DataFrame with the mean house values by neighbourhood per year

df_average_value = melb_stats2.groupby("Suburb").mean().sort_values("Price", ascending=False)
df_average_value["Price per room"] = df_average_value["Price"]/df_average_value["Rooms"]
df_average_value = df_average_value.sort_values("Price per room", ascending=False).head(30)
```

```python
df_average_value["Price per room"].plot(kind="bar", figsize = (30,20), title = "MELBOURNE TOP 30 $/ROOM")
```
