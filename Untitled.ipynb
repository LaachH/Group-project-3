```python
# Helper create_bar_chart function
def create_bar_chart(data, title, xlabel, ylabel, color):
    """
    Create a barplot based in the data argument.
    """
    
    fig = plt.figure()
    bar_chart = melb_stats2.plot.bar(color=color)
    bar_chart.set_xlabel(xlabel, fontsize = 10)
    bar_chart.set_ylabel(ylabel, fontsize = 10)
    bar_chart.set_title(title, fontsize = 12, fontweight = 'bold')
    plt.show()
    plt.Price(fig)
```

```python
# Create a bar chart per year to show the number of dwelling types

# Bar chart for 2017
create_bar_chart(melb_stats2, "Me1bourne prices", "2017", "Dwelling Unit Type", "red")



```
